### Downloading pages and following links are the basis of many web crawling programs.  Python program to backup an entire web site by following all its links.

#### Solution:

```python
import requests
import bs4
```

#### Get URLs from https://www.amazon.in/

```python
res = requests.get('https://www.amazon.in/')
res.raise_for_status()
amritaSoup = bs4.BeautifulSoup(res.text,"lxml")
type(amritaSoup)

file = open("Amazonfile.txt","w") 

for a in amritaSoup.find_all('a',href=True):
    #print ("Found the URL:", a['href'])
    file.writelines('URL: '+ 'https://www.amazon.in/'+ a['href']+'\n') 

file.close()
```

#### Save to Amazonfile.txt

```python
file_in = open("Amazonfile.txt","r")  
for line in file_in:
    
    print(line)
```

#### Get URL related to mobile: 

```python
file_in = open("Amazonfile.txt","r")  
page = []
for line in file_in:
    if ('mobile') in line:
        print(line)
        page.append(line)
```

```python
page[0].strip('URL: \n')
```

#### Extracts all the URLs from web page 'page':

```python
url = page[0].strip('URL: \n')
response = requests.get(url)
 
data = response.text
soup = bs4.BeautifulSoup(data, 'lxml')
tags = soup.find_all('a')
for tag in tags:
    print('URL: '+  str(tag.get('href')))
```
